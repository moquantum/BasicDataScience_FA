<div class = "norm_fa">
دستور آخری که برای ایجاد و تنظیم نمودارهای داخل یک شکل بررسی میکنیم دستور subplot2gridه که شاید کار باهاش از دستور axes آسونتر باشه. این دستور دو تا tuple به عنوان ورودی میگیره که اولی مشخص میکنه که چند سطر و چند ستون نمودار ایجاد بشه. دومی مشخص میکنه ما کدوم سطر و ستون مد نظرمونه. اما دو تا پارامتر دیگه به نامهای rowspan و colspan هم میتونیم بهش بدیم که تعیین میکنه اون نموداری که سطر و ستونش رو مشخص کردیم چند سطر یا چند ستون کش بیاد:
</div>

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3) # ایجاد 3 سطر و 3 ستون نمودار، شروع از سطر 0 ستون 0 ، کش دادن نمودار به اندازه 3 ستون
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2) # شروع از سطر 1 ستون 0 ، کش دادن نمودار به اندازه 2 ستون
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2) # شروع از سطر 1 ستون 2 ، کش دادن نمودار به اندازه 2 سطر
ax4 = plt.subplot2grid((3, 3), (2, 0))# شروع از سطر 2 ستون 0 و نمودار به اندازه عادی
ax5 = plt.subplot2grid((3, 3), (2, 1))# شروع از سطر 2 ستون 1 و نمودار به اندازه عادی
```

اگر به یک جای دلخواه نمودار بخواهیم متن اضافه کنیم میتونیم از تابع text استفاده کنیم. پارامتر اول و دوم این تابع جای متن رو به صورت درصد از عرض و ارتفاع اون نمودار میگیرن و پارامتر سوم متن برای نوشتن رو مشخص میکنه. با دو پارامتر اختیاری va و ha میشه حالت چینش (Alignment) افقی و عمودی متن رو تعیین کرد و پارامترهای اختیاری fontsize و rotation رو هم میشه برای تغییر اندازه متن و چرخش اون استفاده کرد. حالا اگر بخواهیم وسط نمودارهای مثال قبلی یک متن بنویسم اینجوری میشه: 
</div>

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.text(0.5, 0.5, "ax1 (0,0) colspan = 3", va="center", ha="center")# با این حالت چینش، مرکز متن در نقطه مشخص شده خواهد بود
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax2.text(0.5, 0.5, "ax2 (1,0) colspan = 2" ) # اگر حالت چینش را مشخص نکنیم متن از نقطه اعلام شده شروع میشود
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax3.text(0.5, 0.5, "ax3 (1,2)", ha="center", fontsize = 18) # اندازه فونت
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax4.text(0.5, 0.5, "ax4 (2,0)")
ax5 = plt.subplot2grid((3, 3), (2, 1))
ax5.text(0.5, 0.5, "ax5 (2,1)", va="center", ha="center", rotation = -90) #چرخش متن به درجه
```

<div class = "norm_fa">
بعضی اوقات نیاز داریم که دو نمودار رو روی هم یا کنار هم رسم کنیم که یکی از محورهاشون با هم مشترکه ولی محدوده محور دیگشون خیلی با هم متفاوته. در این مواقع توابع twinx و twiny به کمک ما میان. مثلا دو تابع سینوسی و نمایی که محور x شون با هم مشترک باشه:
</div>

```python
plt.figure()
y4 = np.exp(x);
ax1 = plt.gca() # شکل فعال
ax2 = ax1.twinx() # شکل جدید با محور افقی مشترک
ax1.plot(x,y1, color = 'red')
ax1.set_ylabel('sin(x)', color = 'red') # متن و رنگ برچسب محور عمودی
ax1.tick_params(axis='y', labelcolor='red') # رنگ برچسب تیکهای محور عمودی
ax2.plot(x,y4, color = 'blue')
ax2.set_ylabel('exp(x)', color = 'blue')
ax2.tick_params(axis='y', labelcolor='blue')
```

<div class = "norm_fa">
در مثال بالا از تابع set_ylabel استفاده کردیم که علاوه بر متن برچسب محور، رنگ  و البته چیزهای دیگه محور رو میتونه عوض کنه. برای تغییر رنگ بر چسب تیکهای محورهای عمودی هم از تابع tick_params استفاده کردیم.
</div>

<div class = "norm_fa">
قبلا با توابع xlim و ylim که محدوده محورهای (axis) x , y رو تعیین میکردن آشنا شدیم.
با استفاده از توابع xticks و  yticks میتونیم مقادیری که روی محورها نشون داده میشن رو عوض کنیم. با پارامتر rotation هم میتونیم برچسب ها رو بچرخونیم. مثلا اگر روی محور (axis) x، مقادیر pi و pi- برای ما مهم هستند:
</div>

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1], rotation = 90);
```

<div class = "norm_fa">
میتونیم روی اون نقاطی که در دو تابع بالا مشخص کردیم، برچسبهای دلخواه بگذاریم. این برچسب ها از <a href="https://fa.wikipedia.org/wiki/%D9%84%D8%A7%D8%AA%DA%A9">LaTeX</a> پشتیبانی میکنن. مثلا اگر در محور x به جای 3.14 بخواهیم بنویسیم pi:
</div>

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
```

<div class = "norm_fa">
بعضی اوقات یک نقطه روی محور x یا روی محور y برای ما مهم هستند و میخواهیم اون رو مشخص کنیم. برای اینکار میتونیم یک خط عمودی یا افقی رو اون عدد مهم رسم کنیم. برای خط افقی از تابع axhline استفاده میکنیم که یک نقطه روی محور y میگیره و به موازات محور x یک خط رسم میکنه. برای خط عمودی از تابع axvline میشه استفاده کرد. هر دوی این توابع یک پارامتر color میگیرن که رنگ خط رو مشخص میکنه و یک پارامتر lw که ضخامت خط رو تنظیم میکنه:
</div>

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.axhline(0)
plt.axvline(0, color = 'red', lw =2)
```

<div class = "norm_fa">
تا حالا تمام شکل های ما محورهای عمودی و افقی رو سمت پایین و چپ نمودار رسم میکنند. اگر بخواهیم جای رسم محورها رو جابجا کنیم یا رنگ اون رو عوض کنیم از دیکشنری spines در اون نمودار استفاده میکنیم و توابع setcolor و setposition اون شیء رو فراخوانی میکنیم:
</div>

```python
plt.plot(x,y1)
plt.plot(x,y2)
ax = plt.gca()
ax.spines['right'].set_color('none') # محور سمت راست مدرج نیست. مخفیش میکنیم
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0)) # محور پایین مدرج شده محور افقی است. جابجاش میکتیم
ax.spines['left'].set_position(('data',0))
```

<div class = "norm_fa">
برای درجه بندی های (ticks) روی محورها هم میشه تنظیماتی انجام داد. با فراخوانی تابع set_major_locator روی محور نمودار میشه جای درجه ها رو جابجا کرد. مثالهایی از پارامترهایی که میشه به این تابع داد در شکل زیر اومدن: 
</div>

<div class = "norm_fa">
یک مورد مهم دیگه وقتیه که پارامترهای یک محور به جای عدد از جنس زمان باشن:
</div>

```python
import matplotlib.dates as mdates
from datetime import datetime
time = [
    '2019-09-01','2019-09-02','2019-09-03','2019-09-04','2019-09-05','2019-09-06','2019-09-07','2019-09-08','2019-09-09','2019-09-10',
    '2019-09-11','2019-09-12','2019-09-13','2019-09-14','2019-09-15','2019-09-16','2019-09-17','2019-09-18','2019-09-19','2019-09-20',
    '2019-09-21','2019-09-22','2019-09-23','2019-09-24','2019-09-25','2019-09-26','2019-09-27','2019-09-28','2019-09-29','2019-09-30',
    '2019-10-01','2019-10-02'
    ]
time = [datetime.strptime(d, "%Y-%m-%d") for d in time]
plt.figure(figsize=(14,6))
plt.plot(time,y1)
plt.plot(time,y2)
ax = plt.gca()
ax.get_xaxis().set_major_locator(mdates.DayLocator())
ax.get_xaxis().set_major_formatter(mdates.DateFormatter("%Y/%m/%d"))
plt.xticks(rotation=70)
```

<div class = "norm_fa">
برای مشخص کردن و توضیح دادن روی یک نقطه نمودار از تابع annotate استفاده میکنیم. ورودیهای این تابع یک متنه (که از LaTeX هم پشتیبانی میکنه)، یک توپل به اسم xy که طول عرض نقطه ای رو که میخواهیم بهش اشاره کنیم مشخص میکنه، یک پارامتر توپل به نام xytext که طول و عرض جایی که متن نوشته میشه رو تعیین میکنه و یک دیکشنری به نام arrowprops که مشخصات اون پیکان متصل کننده متن و نقطه رو عوض میکنه:
</div>

```python
plt.plot(x,y1)
plt.plot(x,y2)
ax = plt.gca()
ax.annotate('cos(0) = 1', xy=(0, 1), xytext=(1, 1.5), arrowprops=dict(arrowstyle="->"))
ax.set_ylim(-2, 2)
```
